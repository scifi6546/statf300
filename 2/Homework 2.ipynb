$$ P(A \cup B) = P(A) + P(B) - P(A\cap B) = 0.6 + 0.4 - 0.3 = 0.7 $$

# 24 Show that if one event A is contained in another event B (i.e., A is a subset of B), then $$ P(A) \leq P(B) $$ . \[Hint: For
such A and B, A and  $$ B \cup A' $$ are disjoint and $$ B =

general A and B, what does this imply about the relation-
ship among $$ P(A \cup B)$$, $$P(A)$$ and $$P(A \cup B)$$?

If A is contained in B then
$$
P(B) = P(A \cup B) = P(A) + P(B) - P(a \cap B) \Rightarrow P(B) = P(A) + P(B) - P(a \cap B)

## a. If it has an emergency locator, what is the probability that it will not be discovered?
$$

$$
